```python
import os
import json

import numpy as np
from scipy.io import arff
import pandas as pd

from sklearn.model_selection import train_test_split
```

```python
ds_name = 'climate'
```

```python
example = arff.loadarff(os.path.join('new_datasets', f'{ds_name}.arff'))
df = pd.DataFrame(example[0])
df
```

```python
class_col = df.columns[-1]
class_col
```

```python
# convert the binary columns to str
decoder = lambda x: x.decode('utf-8')

for col in df.columns:
    # if df[col].dtype == 'object':
    #     df[col] = df[col].apply(decoder)
    try:
        df[col] = df[col].apply(decoder)
    except:
        pass
df
```

```python
for col in df.columns:
    if df[col].dtype == 'object':
        print(col, df[col].value_counts())
```

```python
df.isna().sum()
```

```python
# convert last column to int by a mapping
df[class_col] = df[class_col].map({k: i for i, k in enumerate(df[df.columns[-1]].unique())})
```

```python
df = pd.get_dummies(df, columns=list(filter(lambda c: df[c].dtype == 'object', df.columns)))
```

```python
df
```

```python
class_labels = df[class_col]
df.drop(columns=[class_col], axis=1, inplace=True)
df = pd.concat([df, class_labels], axis=1)
df
```

```python
df[class_col]
```

```python
df[class_col].value_counts()
```

```python
X, y = df[df.columns[:-1]], df[df.columns[-1]]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2)
```

```python
train = pd.concat([X_train, y_train], axis=1)
test = pd.concat([X_test, y_test], axis=1)
val = pd.concat([X_val, y_val], axis=1)

train.shape, test.shape, val.shape
```

```python
data_path = os.path.join('Tabular', ds_name.capitalize(), 'processed', 'full')
os.makedirs(data_path, exist_ok=True)

train.to_csv(os.path.join(data_path, 'train.csv'), index=True)
test.to_csv(os.path.join(data_path, 'test.csv'), index=True)
val.to_csv(os.path.join(data_path, 'val.csv'), index=True)
```
